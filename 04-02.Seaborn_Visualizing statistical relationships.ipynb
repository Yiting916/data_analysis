# Visualizing statistical relationships
* https://seaborn.pydata.org/tutorial/relational.html

統計分析是了解資料集中的變數如何相互關聯以及這些關係如何依賴其他變數的過程。視覺化可以成為此過程的核心組成部分，因為當資料正確視覺化時，人類視覺系統可以看到表明關係的趨勢和模式。最常用的是 [relplot()](https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot)，用於使用兩種常見方法來視覺化統計關係：散點圖(scatter plots)和線圖(line plots)。
* [scatterplot()](https://seaborn.pydata.org/generated/seaborn.scatterplot.html#seaborn.scatterplot) (with kind="scatter"; the default)
* [lineplot()](https://seaborn.pydata.org/generated/seaborn.lineplot.html#seaborn.lineplot) (with kind="line")

# 使用線圖來顯示連續性
對於某些資料集，您可能想要了解一個變數隨時間的變化，或類似的連續變數。在這種情況下，畫線圖是一個不錯的選擇。在seaborn中，這可以透過lineplot()函數來完成，可以直接使用，也可以透過設定kind="line"來使用relplot()來完成：

```python
dowjones = sns.load_dataset("dowjones")
dowjones.head()
```

```python
# 直線圖畫法 kind = "line"
sns.relplot(data=dowjones, x="Date", y="Price", kind="line")
plt.show()
# 等同於：sns.lineplot
sns.lineplot(x=dowjones['Date'], y=dowjones['Price'])
plt.show()
```

```python
# 如果不使用線圖會變成這樣
sns.relplot(data=dowjones, x="Date", y="Price")
```

# 聚合和表示不確定性
更複雜的資料集將對 x 變數的相同值進行多次測量。 seaborn 中的預設行為是透過繪製平均值和圍繞平均值的 95% 置信區間來聚合每個 x 值的多個測量值：

```python
# 核磁共振的資料
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.relplot(data=fmri, x="timepoint", y="signal", kind="line")
```

```python
#練習題
#使用下面資料，仿造上圖來繪圖
taxis = sns.load_dataset("taxis")
#taxis.head()
sns.relplot(taxis, x="passengers", y="tip", kind="line")
```

```python
# 如果不使用線圖會變成這樣
sns.relplot(data=fmri, x="timepoint", y="signal")
```

置信區間是使用自舉法計算的，對於較大的資料集來說，這可能非常耗時。因此可以禁用它們：

```python
# errorbar -> 設置信任(信心)區間, 預設是95%
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", errorbar=None,
)
```

另一個不錯的選擇，特別是對於較大的數據，是透過繪製標準差而不是置信區間來表示每個時間點的分佈分佈：

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", errorbar="sd",
)
```

# Showing multiple relationships with facets
雖然函數可以同時顯示多個語義變量(資料欄位)，但這樣做並不總是有效。但是，當您確實想了解兩個變數之間的關係如何依賴多個其他變數時該怎麼辦？

最好的方法可能是製作多個圖。因為 relplot() 是基於 FacetGrid，所以這很容易做到。為了顯示附加變數的影響，不要將其指派給圖中的語義角色之一，而是使用它來「分面」視覺化。這意味著您創建多個軸並在每個軸上繪製資料子集：

```python
tips.head()
```

```python
# 在一張圖中分別使用顏色跟點樣式來代表smoker與time
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", style="time",
)
```

↑ 上面這張圖雖然分別使用顏色跟樣式來代表不同的資料欄位，但是看起來還是很雜亂<BR>
↓ 比較合適的方法是將 time 這個欄位分成兩張不同的圖來並列顯示

```python
# 用 time 欄位來切圖
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", col="time",
)
```

```python
#練習題
#使用下面資料，仿造上圖來繪圖
penguins = sns.load_dataset("penguins")
#penguins.head()

# 顯示島嶼上的企鵝品種分布圖
sns.relplot(penguins, x="bill_length_mm", y="body_mass_g", hue="species", col="island", style="sex")
```

您也可以透過這種方式顯示兩個變數的影響：一種透過在列上分面，一種透過在行上分面。當您開始在網格上新增更多變數時，您可能需要減少圖形大小。請記住，FacetGrid 的大小由每個面的高度和縱橫比參數化：

```python
sns.relplot(
    data=tips,
    x="total_bill", y="tip", hue="smoker", col="time", row="sex"
)
```

```python
#練習題
#使用下面資料，仿造上圖來繪圖
penguins = sns.load_dataset("penguins")
#penguins.head()
# 顯示島嶼上的企鵝品種和性別的分布圖
sns.relplot(penguins, x="bill_length_mm", y="body_mass_g", hue="species", col="island", row="sex")
```

```python
# 另一個範例
fmri.head()
```

```python
sns.relplot(
    data=fmri, kind="line",
    x="timepoint", y="signal", hue="subject",
    col="region", row="event", height=3,
    estimator=None
)
```

當您想要檢查變數的多個層級的影響時，最好在列上對該變數進行分面，然後將分面「包裝(wrap)」到行中：

```python
# 不使用col_wrap時的效果
sns.relplot(
    data=fmri.query("region == 'frontal'"), kind="line",
    x="timepoint", y="signal", hue="event", style="event",
    col="subject", #col_wrap=5,
    height=3, aspect=.75, linewidth=2.5,
)
```

```python
sns.relplot(
    data=fmri.query("region == 'frontal'"), kind="line",
    x="timepoint", y="signal", hue="event", style="event",
    col="subject", col_wrap=5,
    height=3, aspect=.75, linewidth=2.5,
)
```
