```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("../data/vgsales.csv")
data.head()
```

```python
# 找出有缺失值的欄位
data.isnull().sum()
```

```python
# 計算欄位的基本統計數據
data.describe(include='all')
```

```python
# 取出 "Publisher" 欄位的資料
data['Publisher']
```

```python
# 取出 Publisher, Name, Genre 欄位的資料
data[['Publisher','Name','Genre']]
```

```python
# 計算每筆資料的 NA_Sales,EU_Sales,JP_Sales,Other_Sales 總和
data['NA_Sales'] + data['EU_Sales'] + data['JP_Sales'] + data['Other_Sales']
```

```python
# 計算每筆資料的 NA_Sales,EU_Sales,JP_Sales,Other_Sales 總和，並存入新的欄位"總和"
data['總和'] = data['NA_Sales'] + data['EU_Sales'] + data['JP_Sales'] + data['Other_Sales']
data
```

```python
#找出發行商為Nintendo的遊戲名稱
data[data['Publisher'] == 'Nintendo']['Name']
```

```python
#列出所有Wii的遊戲名稱
data[data['Platform'] == 'Wii']['Name']
```

```python
#計算所有GBA遊戲的銷售總合
data[data['Platform'] == 'GBA']['Global_Sales'].sum().item()
```

```python
#計算任天堂運動類遊戲的銷售總量
data[(data['Publisher'] == 'Nintendo') & (data['Genre'] == 'Sports')]['Global_Sales'].sum().item()
```

```python
#列出北美地區銷售量最高的前五名遊戲
data.sort_values(by='NA_Sales', ascending=False)[0:5]
```

```python
#列出北美與歐洲地區銷售量總和最高的前五名遊戲
data['總和'] = data['NA_Sales'] + data['EU_Sales']
data.sort_values(by='總和', ascending=False)[0:5]
```

```python
#找出總共有多少種不同的平台
data['Platform'].unique()
```

```python
#寫程式計算不同平台的總銷售量
platform = data['Platform'].unique()
data2 = pd.DataFrame() 
data2['Platform'] = platform
sum_temp = []
for p in platform:
    sum_temp.append(data['Global_Sales'][data['Platform'] == p].sum())
data2['Total_Global_Sales'] = sum_temp
data2
```

```python
# 使用GroupBy來完成上述的工作
data.groupby('Platform')['Global_Sales'].sum()
```
