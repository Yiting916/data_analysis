# DataFrame
二維數據結構，類似於 Excel 表格或 SQL 數據表，是 Pandas 中最常用的數據結構。它由多個 Series 組成，每個 Series 代表一列。

```python
population_dict = {'California': 38332521,
                   'Texas': 26448193,
                   'New York': 19651127,
                   'Florida': 19552860,
                   'Illinois': 12882135}
population = pd.Series(population_dict, name="人口")
print(population)
print(" ")
area_dict = {'California': 423967, 'Texas': 695662, 'New York': 141297, 'New York2': 141297,
             'Florida': 170312, 'Illinois': 149995}
area = pd.Series(area_dict, name="面積")
print(area)
```

```python
states = pd.DataFrame({'人口': population,
                       '面積': area})
states
```

```python
states.index
```

```python
states.columns
```

```python
# 練習題
# 請建立另一個體重與姓名關係的Series
# 並再建立一個包含 體重與身高 的 DataFrame,使用姓名當索引
height_dict = {'David':162, 'Ken':169, 'Sam': 173, 'Owen':165, 'Paul':177}
height = pd.Series(height_dict, name="身高(cm)")
#print(height)
weight_dict = {'David':70, 'Owen':80, 'Paul':78, 'Ken':65, 'Sam': 83}
weight = pd.Series(weight_dict, name="體重(kg)")
#print(weight)
person = pd.DataFrame({'身高': height, '體重': weight})
person
```

```python
#取得特定欄位的資料
states['面積']
```

```python
# 檢查是否包含某個索引
'Texas' in states.index
```
